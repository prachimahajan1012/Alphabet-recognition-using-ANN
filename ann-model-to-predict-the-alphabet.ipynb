## ***Loading Dataset***

```python
df = pd.read_csv("../input/az-handwritten-alphabets-in-csv-format/A_Z Handwritten Data.csv").astype('float32')
df.rename(columns={'0':'label'}, inplace=True)
```

```python
X= df.drop('label',axis = 1)
Y= df['label']
```

## ***Splitting data in train & test***

```python
from sklearn.model_selection import train_test_split

#Splitting dataset in 80% and 20% = train & test data
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
#297960 images , 28*28 = 748 pixels
X_train.shape , y_train.shape
```

## ***Standardization using MinMaxScaler***

```python
#scaling the data with MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
Scaler.fit(X_train)
X_train = Scaler.transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
#we have 26 alphabets , we need 26 values for y train and y test. 
from keras.utils import np_utils
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
y_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow import keras
```

```python
X_train.shape[1]
```

## ***Making Sequential Model & adding layers***

```python
model = keras.Sequential()

model.add(layers.Dense(400, activation="relu" , input_dim = X_train.shape[1])) 
model.add(layers.Dense(300, activation="relu"))
model.add(layers.Dense(300, activation="relu"))
model.add(layers.Dense(300, activation="relu"))
model.add(layers.Dense(200, activation="relu"))
model.add(layers.Dense(100, activation="relu"))
model.add(layers.Dense(len(Y.unique()), activation= "softmax"))
```

## ***Defining Optimizer & Compiling Model***

```python
adam = keras.optimizers.Adam(learning_rate=0.001,decay=1e-6)
model.compile(loss='categorical_crossentropy', optimizer=adam , metrics = ['Accuracy'])
```

## ***Fitting model on training data***

```python
model.fit(X_train,y_train,epochs=20,batch_size=32)
```

## ***Predicting values***

```python
model.predict(X_train[0].reshape(1,784))
```

```python
model.predict(X_train[0].reshape(1,784)).argmax()
```

## ***Function to get output for a given input image***

```python
def predict_alphabet(x):
    x = model.predict(x.reshape(1,784)).argmax()
    dictionary = {'a':1,'b':2,'c':3,'d':4,'e':5,'f':6,'g':7,'h':8,
    'i':9,'j':10,'k':11,'l':12,'m':13,'n':14,'o':15,'p':16,'q':17,
    'r':18,'s':19,'t':20,'u':21,'v':22,'w':23,'x':24,'y':25,'z':26
      }
    for key, value in dictionary.items():
        if value == x:
            print(key)
```
